 ## Преобразование типа данных в столбцах

###  Функция преобразования

```python
def convert_to_datetime(df, columns):
    """
    Функция для приведения указанных столбцов датафрейма к типу данных "datetime".

    Аргументы:
    df: pandas DataFrame
        Датафрейм, в котором нужно привести столбцы к типу данных "datetime".
    columns: list of str
        Список названий столбцов, которые необходимо привести к типу данных "datetime".

    Возвращает:
    Ничего не возвращает. Преобразует столбцы датафрейма к типу данных "datetime".
    """
    for column in columns:
        df[column] = pd.to_datetime(df[column], unit='s')
```

```python
def gradient_color(value):
    """
    Функция преобразует числовое значение в строку CSS для задания градиента цвета.

    Параметры:
        value (float): Числовое значение, которое определяет интенсивность цвета.

    Возвращает:
        str: Строка CSS с заданным цветом в формате 'background-color: #RRGGBB', где RR, GG, BB - шестнадцатеричные значения цвета.

    Примечание:
        Функция используется для создания градиента цвета в зависимости от значения.
    """
    if value == 0:
        return None
    else:
        intensity = min(1, value / 100)
        color_code = '#{:02x}{:02x}{:02x}'.format(255, int(255 * (1 - intensity)), int(255 * (1 - intensity)))
        return f'background-color: {color_code}'

def percent_finder(df):
    """
    Функция вычисляет процентное количество пропущенных значений в каждом столбце DataFrame и возвращает стилизованный DataFrame с градиентным цветом для визуализации процентов.

    Параметры:
        df (pandas.DataFrame): Исходный DataFrame для анализа пропущенных значений.

    Возвращает:
        pandas.io.formats.style.Styler: Стилизованный DataFrame с примененным градиентным цветом для визуализации процентов пропущенных значений.

    Примечание:
        Функция преобразует процентное количество пропущенных значений в стилевой объект DataFrame с примененным градиентным цветом для визуализации.
    """
    percent = 100*(df.isna().sum() / len(df.index))
    return percent.to_frame().style.map(gradient_color)

```

### Класс

```python

class DataAnalyzer:
    def __init__(self, dataframe):
        self.dataframe = dataframe
    def update_dataframe(self, new_dataframe):
        self.dataframe = new_dataframe

    def unique_values_info(self, column_name):
        """
    Функция выводит информацию о уникальных значениях в указанном столбце DataFrame.

    Параметры:
        column_name (str): Название столбца для анализа уникальных значений.

    Возвращает:
        None

    Примечание:
        Функция выводит количество уникальных записей и список уникальных значений в указанном столбце DataFrame.
        """
        unique_values = self.dataframe[column_name].dropna().unique()
        unique_values_sorted = sorted(unique_values)
        unique_count = len(unique_values_sorted)

        print(f"Количество уникальных записей в столбце '{column_name}': {unique_count}")
        print()
        print(f"Уникальные значения в столбце '{column_name}': {unique_values_sorted}")

    def plot_top_values(self, column_name, group_name, top, title,filter_1=None):
        """
    Функция строит горизонтальную столбчатую диаграмму для топ-значений в указанном столбце DataFrame.

    Параметры:
        column_name (str): Название столбца для анализа.
        group_name (str): Название столбца, по которому будет производиться агрегация.
        top (int): Количество топ-значений для отображения.
        title (str): Заголовок диаграммы.
        filter_1 (str, optional): Значение для фильтрации данных. По умолчанию None.

    Возвращает:
        None

    Примечание:
        Функция строит горизонтальную столбчатую диаграмму для топ-значений в указанном столбце DataFrame.
        Можно использовать параметр filter_1 для исключения определенного значения при построении диаграммы.
        """
    # Фильтрация данных, исключая значения "нет информации"
        if filter_1:
            self.dataframe= self.dataframe[self.dataframe[column_name] != filter_1]

        grouped_df = self.dataframe.groupby(column_name).agg({group_name: 'nunique'}).reset_index()

        sorted_df = grouped_df.sort_values(by=group_name, ascending=False)
        top_df = sorted_df.head(top).sort_values(by=group_name, ascending=True)

        plt.figure(figsize=(8, 5))
        plt.barh(top_df[column_name], top_df[group_name])
        plt.xlabel('Количество')
        plt.ylabel(None)
        plt.title(f'Топ {top} значений для {title}')
        plt.show()

    def plot_line_graph(self, column, id_column, title):
        """
        Визуализирует линейный график зависимости количества уникальных id в указанном столбце DataFrame по месяцам.

        Аргументы:
        column (str): Название столбца, содержащего даты.
        id_column (str): Название столбца с уникальными идентификаторами.
        title (str): Заголовок для графика.
        """
        # Преобразование столбца с датами в формат datetime, если он не является таковым
        self.dataframe = self.dataframe.copy()
        self.dataframe[column] = pd.to_datetime(self.dataframe[column])
        self.dataframe['month'] = self.dataframe[column].dt.strftime('%Y-%m')

        # Группировка данных по месяцу и подсчет количества уникальных id в указанном столбце
        grouped_df = self.dataframe.groupby('month')[id_column].nunique().reset_index(name='count')

        # Построение графика
        plt.figure(figsize=(18,6))
        plt.plot(grouped_df['month'], grouped_df['count'], label=f'Уникальные {id_column}')

        plt.xlabel('Месяц')
        plt.ylabel('Количество')
        plt.title(f'График линейной зависимости {title} по месяцам')
        plt.xticks(rotation=90)  # Поворот подписей по оси X для удобства
        plt.legend()
        plt.show()


    def plot_histogram(self, column_name, bins, Title, comparison_operator=None, comparison_value=None):
        """
    Функция строит гистограмму для указанного столбца DataFrame.

    Параметры:
        column_name (str): Название столбца для построения гистограммы.
        bins (int): Количество интервалов (корзин) для построения гистограммы.
        Title (str): Заголовок гистограммы.
        comparison_operator (str, optional): Оператор сравнения для фильтрации данных (например, '>', '<=', '=='). По умолчанию None.
        comparison_value (int/float, optional): Значение для сравнения при фильтрации данных. По умолчанию None.

    Возвращает:
        None

    Примечание:
        Функция строит гистограмму для указанного столбца DataFrame с заданным количеством интервалов.
        Если указаны оператор сравнения и значение, то данные фильтруются перед построением гистограммы.
        """

        df = self.dataframe.copy()
    
        if comparison_operator and comparison_value is not None:
            condition = f"df['{column_name}'] {comparison_operator} {comparison_value}"
            df = df[eval(condition)]
    
        plt.figure(figsize=(8, 5))
        plt.hist(df[column_name], bins=bins, color='skyblue', edgecolor='black')
        plt.title(f'Гистограмма для {Title}')
        plt.xlabel(column_name)
        plt.ylabel('Частота')
        plt.grid(axis='y', alpha=0.75)
        plt.show()

    def plot_boxplot(self, column_name):
        """
        Функция строит ящик с усами для указанного столбца DataFrame.
    
        Параметры:
            column_name (str): Название столбца для построения ящика с усами.
    
        Возвращает:
            None
    
        Примечание:
            Функция строит ящик с усами для указанного столбца DataFrame.
        """
    
        plt.figure(figsize=(6, 3))
        plt.boxplot(self.dataframe[column_name], vert=False)
        plt.title(f'Ящик с усами для {column_name}')
        plt.xlabel(column_name)
        plt.show()
    def plot_top_values_pie(self, column_name, group_name, Title,top):
        """
        Визуализирует долю топ значений по указанным столбцам в виде круговой диаграммы.

        Аргументы:
        column_name (str): Название столбца, по которому будет проводиться анализ.
        group_name (str): Название столбца с группирующими данными.
        top (int): Количество топ значений для отображения.
        """
        grouped_df = self.dataframe.groupby(column_name).agg({group_name: 'count'}).reset_index()

        # Сортировка по убыванию суммы значений
        sorted_df = grouped_df.sort_values(by=group_name, ascending=False)

        # Выбор топ значений
        top_df = sorted_df.head(top)

        # Построение круговой диаграммы
        plt.figure(figsize=(6, 5))
        patches, texts, autotexts = plt.pie(top_df[group_name], labels=top_df[column_name],
                                            autopct=lambda p: f'{p:.1f}%\n({int(p * sum(top_df[group_name]) / 100)})',
                                            startangle=140, textprops={'fontsize': 10})

        # Установка размера шрифта для текста внутри секторов
        for text in autotexts:
            text.set_fontsize(10)

        plt.title(f'Доля топ {top} значений для {Title}')
        plt.show()
```

```python
def find_group(category,groups):
    """ Функция ищет группу, к которой относится заданная категория, на основе списка ключевых слов.
    Параметры:
        category (str): Категория, для которой нужно найти соответствующую группу.
        groups (dict): Словарь, содержащий группы и их соответствующие ключевые слова.
    Возвращает:
        str: Название группы, к которой относится категория.
    Примечание:
        Функция проходит по каждой группе в словаре groups и сравнивает каждое ключевое слово внутри группы с категорией.
        Если хотя бы одно ключевое слово из группы содержится в категории, функция возвращает название этой группы.
        Если соответствие не найдено, функция возвращает 'Другие услуги'."""
      
    for group, keywords in groups.items():
        for keyword in keywords:
            if keyword in category:  # Поиск с учетом регистра
                return group
    print(f"Не удалось найти соответствие для категории '{category}'")
    return 'Другие услуги'
```

### Метод локтя

**Метод локтя:**
Обоснование применения:1. 

Количественная оценка оптимального числа кластеров: Метод локтя предоставляет количественную оценку оптимального числа кластеров, основанную на изменении инерции с увеличением количества кластеров. Это позволяет принимать обоснованные решения на основе данных2. .

Простота реализации: Реализация метода локтя относительно проста, и его легко применить с использованием стандартных библиотек Python. Это делает метод доступным для широкого круга пользователей без необходимости в глубоких знаниях математики и статистики.

```python
inertias = []

# Проводим кластеризацию для разных k и сохраняем инерцию
k_values = range(1, 11)

# Инициализируем список для сохранения инерций для каждого k
inertias = []

# Проводим кластеризацию для разных k и сохраняем инерцию
for k in k_values:
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)
    kmeans.fit(X)
    inertias.append(kmeans.inertia_)

# Строим график метода локтя
plt.plot(k_values, inertias, marker='o')
plt.xlabel('Количество кластеров')
plt.ylabel('Inertia')
plt.title('Метод локтя для количетва кластеров')
plt.show()
```

### метод силуэта

**Метод силуэта:**


Обоснование применения:1. 

Комплексная оценка качества кластеризации: Метод силуэта предоставляет комплексную оценку качества кластеризации, учитывая как внутрикластерные, так и межкластерные расстояния. Это позволяет учитывать не только компактность кластеров, но и их разделимость2. .

Учет сбалансированности кластеров: Силуэт учитывает сбалансированность кластеров, что позволяет обнаруживать не только компактные, но и равномерно распределенные кластеры. Это особенно важно при работе с данными, содержащими кластеры различных форм и размеров.

```python
# Инициализируйте список для хранения оценок силуэта для каждого k
silhouette_scores = []
# Проведите кластеризацию для различных значений k и рассчитайте оценку силуэта
k_values = range(2, 11)

for k in k_values:
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)
    kmeans.fit(X_scaled)
    labels = kmeans.labels_
    silhouette_avg = silhouette_score(X_scaled, labels)
    silhouette_scores.append(silhouette_avg)

# Plot the silhouette scores
plt.plot(k_values, silhouette_scores, marker='o')
plt.xlabel('Количество кластеров')
plt.ylabel('Оценку силуэта')
plt.title('Оценка силуэта для разного количества кластеров')
plt.show()
```

```python
k_values = range(1, 11)
inertias = []
plt.figure(figsize=(11,6))
for k in k_values:
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)
    kmeans.fit(X_scaled)
    inertias.append(kmeans.inertia_)
# Вычислите первую производную от значений инерции
derivative = np.diff(inertias)
# Построение графика метода elbow с использованием первой производной
plt.subplot(1, 2, 1)
plt.plot(k_values, inertias, marker='o')
plt.xlabel('Количество кластеров')
plt.ylabel('Inertia')
plt.title('Метод логтя')

plt.subplot(1, 2, 2)
plt.plot(range(1, len(derivative) + 1), derivative, marker='o', linestyle='--', color='orange')
plt.xlabel('Количество кластеров')
plt.ylabel('Скорость изменения инерции')
plt.title('Первая производная от инерции')

plt.tight_layout()
plt.show()
```

```python
def churn_detection(date, reference_date):
    # Преобразование строки с датой в формат datetime, если она еще не в таком формате
    if isinstance(date, str):
        row_date = datetime.strptime(date, '%Y-%m-%d')

    # Проверка разницы между датами
    delta = reference_date - date

    # Если разница больше 90 дней, считаем, что произошел отток и возвращаем 1, иначе 0
    if delta.days > 90:
        return 1
    else:
        return 0
```

### Выявление стажа

```python
def calculate_experience_in_months(row):
    """
    Функция вычисляет опыт работы в месяцах на основе даты создания и удаления записи.

    Параметры:
        row (pandas.Series): Строка из DataFrame, содержащая информацию о дате создания, удаления и статусе удаления.

    Возвращает:
        int: Опыт работы в месяцах, округленный до целого числа.

    Примечание:
        Если значение в столбце 'deleted_at' равно NaN, то оно заменяется на текущую дату.
        Если запись была удалена ('is_deleted' равно True), опыт работы вычисляется по разнице между датой создания и удаления.
        В противном случае опыт работы вычисляется по разнице между датой создания и текущей датой.
    """
    created_at = row['created_at']
    deleted_at = row['deleted_at']
    is_deleted = row['is_deleted']
    current_date = datetime.now()  # Получаем текущую дату

    if pd.isnull(deleted_at):  # Проверяем, является ли значение deleted_at NaN
        deleted_at = current_date  # Если значение deleted_at NaN, присваиваем ему текущую дату

    if is_deleted:
        experience_days = (deleted_at - created_at).days
    else:
        experience_days = (current_date - created_at).days

    experience_months = experience_days / 30  # Разделим количество дней на количество дней в месяце

    return round(experience_months)
```

```python
# Применим функцию к DataFrame и создадим новый столбец 'experience'
people['experience'] = people.apply(calculate_experience_in_months ,axis=1)
```

```python

def plot_sep_line(data, y_mean, Title, column=None):
    # Преобразование столбца created_at в формат datetime, если он не является таковым
    data['created_at'] = pd.to_datetime(data['created_at'])

    # Извлечение месяца из столбца created_at
    data['month'] = data['created_at'].dt.strftime('%Y-%m')

    if column is not None:
    #  данных по месяцам и столбцу column (если он задан) и вычисление среднего значения y_mean для каждой группы
        grouped_data = data.groupby(['month', column])[y_mean].mean().reset_index()
        plt.figure(figsize=(10, 6))
        for label, df in grouped_data.groupby(column):
            plt.plot(df['month'], df[y_mean], marker='o', linestyle='-', label=f'{column}={label}')
    else:
        plt.figure(figsize=(8, 5))
        grouped_data = data.groupby(['month'])[y_mean].mean().reset_index()
        plt.plot(grouped_data['month'], grouped_data[y_mean], marker='o', linestyle='-', label='All')

    # Настройка осей и заголовка

    plt.xlabel('Month')
    plt.ylabel('Average ' + y_mean)
    plt.title(Title)

    # Добавление легенды
    plt.legend()

    # Отображение графика
    plt.grid(True)
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()
```

```python
def scatterplot_average_experience_and_unique_ids(data):
    # Преобразование столбца created_at в формат datetime, если он не является таковым
    data['created_at'] = pd.to_datetime(data['created_at'])

    # Извлечение месяца из столбца created_at
    data['month'] = data['created_at'].dt.strftime('%Y-%m')

    # Группировка данных по месяцам и подсчет среднего стажа и количества уникальных id для каждого месяца
    grouped_data = data.groupby('month').agg({'experience': 'mean', 'user_id': 'nunique'}).reset_index()

    # Построение scatter plot
    plt.figure(figsize=(10, 6))
    plt.scatter(grouped_data['month'],  grouped_data['experience'], grouped_data['user_id'], marker='o')

    # Настройка осей и заголовка
    plt.xlabel('Даты')
    plt.ylabel('Средний стаж')
    plt.title('Scatter plot: Средний стаж vs Количество уникальных id')
    plt.xticks(rotation=90)
    # Отображение графика
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

 ## класс Analyzer

```python
class Analyzer:
    def __init__(self, dataframe):
        self.dataframe = dataframe

    def update_dataframe(self, new_dataframe):
        self.dataframe = new_dataframe

    def plot_nan_distribution(self, column_name, title):
        """
    Функция строит круговую диаграмму для визуализации распределения заполненных и пропущенных значений в указанном столбце.

    Параметры:
        column_name (str): Название столбца, для которого нужно построить распределение.
        title (str): Заголовок диаграммы.

    Возвращает:
        None

    Примечание:
        Функция получает количество заполненных и пропущенных значений в указанном столбце.
        Затем создает круговую диаграмму, отображающую процентное соотношение заполненных и пропущенных значений.
        """
        # Получаем количество заполненных и NaN значений в столбце
        filled_count = self.dataframe[column_name].count()
        nan_count = self.dataframe[column_name].isnull().sum()

        # Создаем список для подписей диаграммы с процентами и количеством
        labels = [f'Используют подписку\n({filled_count} записей)', f'Не используют подписку\n({nan_count} записей)']

        # Создаем список для значений диаграммы
        sizes = [filled_count, nan_count]

        # Создаем круговую диаграмму
        plt.figure(figsize=(8, 6))
        plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
        plt.title(title)

        # Отображаем диаграмму
        plt.show()

    def plot_category(self, metric, xlabel, title1,Rotation):
        """
        Построение barplot для визуализации распределения категориальной переменной и её влияния на отток.

        Параметры:
        - metric (str): Название категориальной переменной.
        - xlabel (str): Название оси x.
        - title1 (str): Заголовок графика.

        Возвращает:
        None: Отображает график.

        Пример:
        plot_category('название_вашей_категориальной_переменной', 'Название x-оси','Заголовок')
        """
        plt.figure(figsize=(8, 6))

        # Создание barplot с указанием параметра x и y
        ax = sns.barplot(data=self.dataframe, x=metric, y='churn')

        # Добавление линии среднего уровня оттока
        plt.axhline(self.dataframe['churn'].mean(), color='red', linestyle='dashed', linewidth=2, label='Средний отток')

        # Настройка легенды
        plt.legend()

        # Настройка осей и заголовка
        plt.xlabel(xlabel)
        plt.ylabel('Уровень оттока')
        plt.title(title1)
        plt.xticks(rotation=Rotation)
        # Отображение графика
        plt.show()

    def pie(self,column, title,first,second):
        count_1 = self.dataframe[self.dataframe[column] == 1][column].count()
        count_0 = self.dataframe[self.dataframe[column] == 0][column].count()
        labels = [f'{first} ({count_1})', f'{second} ({count_0})']
        sizes = [count_1, count_0]
        plt.figure(figsize=(8, 6))
        plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
        plt.title(title)
        plt.show()
    def _calculate_monthly_counts(self, data, column, first_category, second_category, month_of_year=True):
        """
        Вычисляет ежемесячные значения для заданных категорий столбца DataFrame.

        Параметры:
        - data (DataFrame): Датафрейм, для которого нужно вычислить значения.
        - column (str): Название столбца, для которого нужно вычислить значения.
        - first_category (int): Значение первой категории.
        - second_category (int): Значение второй категории.
        - month_of_year (bool, optional): Флаг, указывающий, нужно ли использовать месяц и год в качестве категории. По умолчанию False.

        Возвращает:
        first_category_monthly_counts (Series): Ежемесячные значения для первой категории.
        second_category_monthly_counts (Series): Ежемесячные значения для второй категории.
        """
        # Copy the dataset to avoid modifying the original data
        df = data.copy()

        if month_of_year:
            # Используем месяц и год в качестве категории
            df['date'] = df['created_at_billing'].dt.strftime('%Y-%m')
        else:
            # Используем только месяц в качестве категории
            df['date'] = df['created_at_billing'].dt.strftime('%Y')

        # Фильтруем данные по категориям
        first_category_data = df[df[column] == first_category]
        second_category_data = df[df[column] == second_category]

        # Группируем данные по месяцам и считаем количество записей для каждой категории
        first_category_monthly_counts = first_category_data.groupby('date').size()
        second_category_monthly_counts = second_category_data.groupby('date').size()

        return first_category_monthly_counts, second_category_monthly_counts

    def plot_monthly_trend(self, column, title, first_category, second_category):
        """
        Функция для отображения динамики по месяцам для заданных категорий столбца DataFrame.

        Параметры:
        - column (str): Название столбца, для которого нужно построить динамику.
        - title (str): Заголовок графика.
        - first_category (int): Значение первой категории.
        - second_category (int): Значение второй категории.

        Возвращает:
        None: Отображает график.

        Пример использования:
        analyzer.plot_monthly_trend('столбец', 'Динамика по месяцам', 1, 0)
        """
        # Вычисляем ежемесячные значения для заданных категорий
        first_category_monthly_counts, second_category_monthly_counts = self._calculate_monthly_counts( self.dataframe, column, first_category, second_category)

        # Вычисляем процентное соотношение от общего количества для каждой категории
        total_counts = first_category_monthly_counts + second_category_monthly_counts
        first_category_percentages = (first_category_monthly_counts / total_counts) * 100
        second_category_percentages = (second_category_monthly_counts / total_counts) * 100

        # Строим график
        plt.figure(figsize=(10, 6))
        plt.plot(first_category_percentages.index, first_category_percentages.values, marker='o', linestyle='-', label=f'{first_category}')
        plt.plot(second_category_percentages.index, second_category_percentages.values, marker='o', linestyle='-', label=f'{second_category}')
        plt.title(title)
        plt.xlabel('Месяц')
        plt.ylabel('Процент от общего количества')
        plt.xticks(rotation=45)
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.show()

    def _highlight_outliers(self, df):
        # Функция для выделения выделяющихся значений в таблице
        styled_df = df.style.apply(lambda x: x.map(gradient_color), axis=None)
        return styled_df



    def plot_monthly_trend_table(self, column, first_category, second_category, column2):
        """Функция для отображения таблицы с процентным соотношением по месяцам для заданных категорий столбца DataFrame.

        Параметры:
        - column (str): Название столбца, для которого нужно построить динамику.
        - first_category (int): Значение первой категории.
        - second_category (int): Значение второй категории.
        - column2 (str): Название столбца, по которому нужно строить таблицы для каждого уникального значения. По умолчанию None.

        Возвращает:
        None: Отображает таблицы.

        Пример использования:
        analyzer.plot_monthly_trend_table('столбец', 1, 0, column2='название_столбца') """

        unique_values = self.dataframe[column2].unique()
        display(f'Информация для: { column2 }')
        for value in unique_values:
            # Фильтруем данные по текущему значению, если column2 не пустой
            if column2:
                value_data = self.dataframe[self.dataframe[column2] == value]
            else:
                value_data = self.dataframe  # Используем весь датафрейм

            # Вычисляем ежемесячные значения для заданных категорий
            first_category_monthly_counts, second_category_monthly_counts = self._calculate_monthly_counts(value_data,column, first_category, second_category,False)

            # Вычисляем процентное соотношение от общего количества для каждой категории
            total_counts = first_category_monthly_counts + second_category_monthly_counts
            first_category_percentages = (first_category_monthly_counts / total_counts) * 100
            second_category_percentages = (second_category_monthly_counts / total_counts) * 100

            # Создаем таблицу
            df_table = pd.DataFrame({f'{first_category}': first_category_percentages.values,
                                     f'{second_category}': second_category_percentages.values},
                                    index=first_category_percentages.index)  # Индексом станет столбец "Месяц"

            # Применяем выделение выделяющихся значений
            df = df_table.fillna(0)
            styled_df = self._highlight_outliers(df.transpose())

            # Отображаем таблицу
            display(f'Столбец {value if column2 else "все значения"}:')
            display(styled_df )
            print()
```
